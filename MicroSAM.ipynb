```python
import tifffile
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# Define the base directory as the current working directory
base_dir = os.getcwd()  # CMP directory

# Intensity and labels
original_intensity = tifffile.imread(os.path.join(base_dir, "result_demo", "intensity-01.tif"))

manually_label = tifffile.imread(os.path.join(base_dir, "result_demo", "manually_label_01.tif"))

# Pretrained SAM 
SAM_seg = tifffile.imread(os.path.join(base_dir, "result_demo", "SAM_seg_01.tif"))

# MicroSAM
point_prompt_seg = tifffile.imread(os.path.join(base_dir, "result_demo", "point_prompt_01.tif"))

box_prompt_seg = tifffile.imread(os.path.join(base_dir, "result_demo", "box_prompt_01.tif"))

auto_seg = tifffile.imread(os.path.join(base_dir, "result_demo", "auto_seg_01.tif"))

auto_rectified = tifffile.imread(os.path.join(base_dir, "result_demo", "auto_rectified_01.tif"))



```

```python
# Number of slices to display
num_slices = 12

# Create a figure with a 6x12 grid of subplots
fig, axes = plt.subplots(7, num_slices, figsize=(16, 7))

for i in range(num_slices):
    # Plot the ith slice of original_intensity
    axes[0, i].imshow(original_intensity[3*i+5], cmap='gray')
    axes[0, i].axis('off')
    axes[0, i].set_title(f'Slice {3*i+5}')

    # Plot the ith slice of manually_label
    axes[1, i].imshow(manually_label[3*i+5], cmap='nipy_spectral')
    axes[1, i].axis('off')

    # Plot the ith slice of auto_rectified
    axes[2, i].imshow(SAM_seg[3*i+5], cmap='nipy_spectral')
    axes[2, i].axis('off')

    # Plot the ith slice of auto_rectified
    axes[3, i].imshow(auto_seg[3*i+5], cmap='nipy_spectral')
    axes[3, i].axis('off')
    
    # Plot the ith slice of auto_rectified
    axes[4, i].imshow(auto_rectified[3*i+5], cmap='nipy_spectral')
    axes[4, i].axis('off')

    # Plot the ith slice of auto_rectified
    axes[5, i].imshow(box_prompt_seg[3*i+5], cmap='nipy_spectral')
    axes[5, i].axis('off')

    # Plot the ith slice of auto_rectified
    axes[6, i].imshow(point_prompt_seg[3*i+5], cmap='nipy_spectral')
    axes[6, i].axis('off')
    
plt.show()


```

Row 1: Original image;    
Row 2: Manually Labelled Masks;   

Row 3: Pretrained SAM;  

Row 4: Automatic segmentation of MicroSAM before rectification  
Row 5: Automatic segmentation of MicroSAM after rectification  
Row 6: using box prompt in MicroSAM  
Row 7: using point prompt in MicroSAM
